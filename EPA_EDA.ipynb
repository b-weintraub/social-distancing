```python
df=pd.read_csv('pollution_us_2000_2016.csv')
df
```

```python
df['Date Local'] = pd.to_datetime(df['Date Local'])
df['Month'] = df['Date Local'].dt.month
df['Year'] = df['Date Local'].dt.year
df['Day'] = df['Date Local'].dt.day
```

```python
df2=df[['Year','Month','Day','State Code','County Code','County','City','State','NO2 Mean']]
df2
```

```python
df3=df.groupby('State')['NO2 Mean'].mean()
df3.plot(kind='bar', figsize=(12,4));
```

```python
df4=df[df['State']=='Washington']
df5=df4.groupby(['Year','Month'])['NO2 Mean'].mean()
df5
```
