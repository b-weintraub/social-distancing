```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data/Beacon-Hill-data-from-May2018.csv',header=1)
df
```

```python
df2=df[['Date','NO2_CAPS']]
df2
```

```python
df3=df2[1:710]  #clean data
df3
```

```python
df3['Date'] = pd.to_datetime(df3['Date'])
df3
```

```python
df3['NO2_CAPS']=df3['NO2_CAPS'].astype(float)
```

```python
df3.info()
```

```python
df3['Day'] = df3['Date'].dt.day
df3['Month'] = df3['Date'].dt.month
df3['Year'] = df3['Date'].dt.year
df3
```

```python
fig,ax= plt.subplots(figsize=(12,2))

ax.plot(df3['Date'],df3['NO2_CAPS'])
ax.set_ylabel('NO2 concentration (ppb)')
ax.set_xlabel('Date')
ax.set_title('NO2 concentrations versus time at Seattle Beacon Hill location')
```
